Question 2.

```python
#a
df_titanic = pd.read_csv("titanic.csv")
print(df_titanic.head())
print(df_titanic.info())
```

```python
#b
mean_age = df_titanic['Age'].mean()

median_fare = df_titanic['Fare'].median()

mode_class = df_titanic['Pclass'].mode().values[0]

print(f"Mean Age: {mean_age}")
print(f"Median Fare: {median_fare}")
print(f"Mode Class: {mode_class}")
```

```python
#c
age_variance = df_titanic['Age'].var()
age_std = df_titanic['Age'].std()

fare_variance = df_titanic['Fare'].var()
fare_std = df_titanic['Fare'].std()

print(f"Age Variance: {age_variance}")
print(f"Fare Standard Deviation: {fare_std}")
```

```python
#d
plt.hist(df_titanic['Age'], bins=20, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')
plt.show()
```

```python
#e
passenger_count = df_titanic.shape[0]


total_fare = df_titanic['Fare'].sum()

print(f"Passenger Count: {passenger_count}")
print(f"Total Fare: {total_fare}")
```

```python
#f
Q1 = df_titanic['Age'].quantile(0.25)
Q3 = df_titanic['Age'].quantile(0.75)
age_iqr = Q3 - Q1

print(f"Age IQR: {age_iqr}")
```

```python
#g
# age
age_skewness = df_titanic['Age'].skew()
age_kurtosis = df_titanic['Age'].kurtosis()

# fare
fare_skew = df_titanic['Fare'].skew()
fare_kurtosis = df_titanic['Fare'].kurtosis()

print(f"Age Skewness: {age_skewness}")
print(f"Age Kurtosis: {age_kurtosis}")
print(f"Fare Skewness: {fare_skew}")
print(f"Fare Kurtosis: {fare_kurtosis}")
```

```python
#h
# 25th percentile of fare
Q1_fare = df_titanic['Fare'].quantile(0.25)

# 75th percentile of fare
Q3_fare = df_titanic['Fare'].quantile(0.75)

print(f"25th Percentile Fare: {Q1_fare}")
print(f"75th Percentile Fare: {Q3_fare}")
```
