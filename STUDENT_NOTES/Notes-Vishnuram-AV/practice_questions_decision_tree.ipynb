Q3. How is a decision tree fit to a dataset. Explain the step-by-step algorithm in terms of the gain function? Explain how the optimal feature (j) and threshold () are chosen at each node during training?

**Answer:**
- **Step-by-step Algorithm**:
    1. Start at the root node with the entire dataset.
    2. For each possible feature `j` and threshold ``, calculate the **gain** of splitting the data.
    3. Select the feature `j` and threshold `` that result in the maximum gain.
    4. Split the dataset into subsets based on the chosen `j` and ``.
    5. Repeat the process recursively for each subset until:
       - A stopping criterion is met (e.g., maximum depth, minimum samples per leaf, etc.).
       - A terminal node (leaf) is reached.
  - **Choosing Optimal Feature and Threshold**:
    - Compute the **gain function** (e.g., information gain or Gini impurity decrease) for each candidate split.
    - Select the feature and threshold that yield the highest gain.

Q6. What are random forests? Compare random forest with bagging and justify which you think is better amongst the two.

**Answer:** Random Forests: An extension of Bagging that selects a random subset of features at each split to decorrelate the trees.

Comparison: Bagging uses all features, which can lead to correlated trees if one feature is dominant. Random Forests force diversity by limiting feature selection.

Verdict: Random Forest is generally better because the decorrelated trees provide a larger reduction in variance.

Q7. Explain how bias and variance change as a function of model complexity using a graph. Describe and indicate the optimal balance point (sweet spot) on the curve where the model achieves the best trade-off between underfitting and overfitting.

**Answer:**

Bias: Decreases as complexity increases (model fits data better).

Variance: Increases as complexity increases (model becomes sensitive to noise).

Sweet Spot: The optimal balance point is the minimum of the Total Error curve, representing the best trade-off between underfitting (high bias) and overfitting (high variance).

**Answer:**

Illustration: A model with low bias and high variance is a complex model (like a deep decision tree) that captures every detail of the training data.
